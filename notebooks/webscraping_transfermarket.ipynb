#### Função

```python
import requests
from bs4 import BeautifulSoup
from ordered_set import OrderedSet
import time

def scrape_page(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find('table', class_='items').find('tbody')
    rows = table.find_all('tr')

    name_player_set = OrderedSet()

    for row in rows:
        td_element = row.find('img')
        if td_element:
            img_url = td_element.get('title')
            name_player_set.add(img_url)

    return name_player_set

def main():
    base_url = "https://www.transfermarkt.com.br/campeonato-brasileiro-serie-a/marktwerte/wettbewerb/BRA1/pos//detailpos/0/altersklasse/alle/plus/1/page/"
    num_pages = 4

    all_name_players = []  # Lista para armazenar todos os resultados

    for page_num in range(1, num_pages + 1):
        url = f"{base_url}{page_num}"
        name_player_set = scrape_page(url)

        # Adicionar os resultados da página à lista global
        all_name_players.extend(name_player_set)

        # Adicionar um atraso de 2 segundo entre as requisições
        time.sleep(2)

    return all_name_players

if __name__ == "__main__":
    all_name_players = main()
```

#### Função

```python
import requests
from bs4 import BeautifulSoup

def scrape_page(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find('table', class_='items').find('tbody')
    rows = table.find_all('td', class_='hauptlink')

    pos_list = []

    for row in rows:
        td_element = row.find_next('td')
        if td_element:
            pos_list.append(td_element.get_text())

    return pos_list

def remove_digits(value):
    return re.sub(r'\d', '', value)


def main():
    base_url = "https://www.transfermarkt.com.br/campeonato-brasileiro-serie-a/marktwerte/wettbewerb/BRA1/pos//detailpos/0/altersklasse/alle/plus/1/page/"
    num_pages = 4

    all_pos = []  # Lista para armazenar todos os resultados

    for page_num in range(1, num_pages + 1):
        url = f"{base_url}{page_num}"
        pos_list = scrape_page(url)

        # Remover dígitos de cada item na lista usando a função remove_digits
        pos_list = [remove_digits(value) for value in pos_list]

        # Filtrar valores não vazios
        pos_list = [value for value in pos_list if value]

        # Adicionar os resultados da página à lista global
        all_pos.extend(pos_list)

        # Adicionar um atraso de 1 segundo entre as requisições
        time.sleep(2)


    return all_pos


if __name__ == "__main__":
    all_pos = main()
    print(all_pos)
```

```python
def get_header_and_body(table):
  header = []
  dados_linhas = []

  if table:
    cabecalho = table.find('thead')
    dados = table.find('tbody')

    if cabecalho:
      header = [th.get_text(strip=True) for th in cabecalho.find_all('th')]

    if dados:
      linhas = dados.find_all('tr')
      for linha in linhas:
        colunas = linha.find_all(['th', 'td'])
        linha_dados = [col.get_text(strip=True) for col in colunas]
                
        if len(linha_dados) < len(header):
          linha_dados.append('') 
                
        dados_linhas.append(linha_dados)

  return header, dados_linhas
```

### Classe ODD HTML

```python
tabela = soup.find_all('table')[1]

rows = tabela.find_all('tr', attrs={'class':'odd'})

data = []

for row in rows: #percorrendo as linhas, pegando os dados..
    data_rows = row.findAll('td')

    # Obter atributos específicos dentro das tags img
    img_data_src = row.find('img', {'class': 'bilderrahmen-fixed'}).get('data-src')
    img_src_flag = row.find('img', {'class': 'flaggenrahmen'}).get('src')
    img_src_club = row.find('img', {'class': ''}).get('src')

    data_rows = [ele.text.strip() for ele in data_rows] # trazendo os textos dentro das tags
    data_row = [ele for ele in data_rows if ele]

    # Adicionar os atributos específicos à lista de dados
    data_row.extend([img_data_src, img_src_flag, img_src_club])
    
    data.append(data_row)

# Criar DataFrame
columns = ['#', 'Jogadores', 'Nac.', 'Idade', 'Clube', 'Melhor marca da carreira', 'Última alteração', 'Valor de mercado', 'Img Data Src', 'Img Src Flag', 'Img Src Club']
df = pd.DataFrame(data, columns=columns)
df
```
