# REGEX (Expressões Regulares)

Regex é uma abreviação para "expressão regular" (do inglês "regular expression"). Trata-se de uma sequência de caracteres que define um padrão de busca em texto. As expressões regulares são utilizadas para realizar operações de busca, extração, substituição e validação de padrões em strings.

As regex são amplamente utilizadas em programação, processamento de texto e em diversas ferramentas e linguagens de programação, como Python, JavaScript, Perl e muitas outras. Com elas, é possível buscar e manipular texto de forma precisa, permitindo a realização de tarefas complexas de forma eficiente.

As expressões regulares são compostas por caracteres literais, como letras e números, e por metacaracteres que possuem um significado especial, como ".", "*", "+", "?" e outros. Esses metacaracteres permitem definir padrões mais flexíveis e poderosos, permitindo encontrar correspondências específicas em um texto.

Por exemplo, uma expressão regular simples como "abc" irá procurar por ocorrências da sequência de caracteres "abc" em um texto. Já uma expressão regular mais complexa, como "[0-9]+", irá buscar por uma ou mais ocorrências de dígitos numéricos em um texto.

As expressões regulares são extremamente versáteis e poderosas, mas também podem ser complexas de se escrever e entender. É necessário aprender a sintaxe e os padrões comuns para tirar o máximo proveito dessa ferramenta.

caractere '.' - Significa tudo, busca qualquer caracter ( exceto nova linha )

caractere '^' - Define o inicio do texto

caractere '$' - Define o final do texto

caractere '' - Funciona como caractere de escape, substitui caracteres especiais para caracteres literais

[] - funciona como 'range', busca padrões nas strings internas do colchetes

0 ou mais ocorrencias na busca , retorna vazio quando não encontrar o padrão .

1 ou mais ocorrencias da expressão anterior
? - determina que a expressão anterior possa ser opcional.

\d - busca por padrões númericos

\D - busca por padrões não numéricos

\s - busca por padrões de espaçamento.

\S - busca qualquer caractere que não seja de espaçamento

\w - busca qualquer caractere de [A-Z e 0-9] | [a-z e 09] e sublinhado

\W - busca qualquer caractere que não seja [A-Z e 0-9] | [a-z e 09] ou sublinhado

{} - definir quantidade de ocorrências da expressão anterior - \d{2} - busca por números com 2 dígitos \d{1-10} busca por numeros de 1 a 10 digitos

| - equivalente ao OU (lógico)

() - usa-se para agrupamento de expressões

re.search() - Método que procura a PRIMEIRA ocorrência do padrão em qualquer posição da string.

re.findall() - Método que procura TODAS ocorrências do padrão em qualquer posição da string.

re.match() - Método que procura a PRIMEIRA ocorrência somente no início da string.

re.compile() - Método que compila um padrão, resulta em um objeto que pode ser utilizado assim como re.

re.sub() - Método que realiza substituições em uma string.

re.split() - Método que separa strings baseadas em padrões regex.

flags
re.DOTALL - faz com que o caractere '.' combine com qualquer padrao, inclusive com uma nova linha, tabulacao

re.IGNORECASE - traz o resultado tanto de maiusculas quanto minusculas

re.MULTILINE faz com que inicio/fim {^, $} considere cada linha.

re.UNICODE faz com que {\w, \W, \b, \B} siga regras de unicode

re.LOCALE faz com que {\w, \W, \b, \B} siga "locale"

re.VERBOSE permite comentários em REGEX.

## Método Search

Permite encontrar a primeira ocorrência de um padrão em uma string.


```python
texto = 'eu tenho 295435454354354545 anos'
```

```python
re.search('\d+',texto)
```

## Método FindAll 

Permite encontrar todas as ocorrências de um padrão de expressão regular em uma string.

```python
texto = 'Meu nome é Felipe , meu nome é Felipe'
```

```python
re.findall('Felipe',texto)
```

## Método Match 

Tenta encontrar a ocorrência de um padrão somente no início de uma string

```python
texto = 'Meu nome é Felipe , meu nome é Felipe'
```

```python
re.match('[A-Z]',texto)
```

## Método Sub

Permite substituir todas as ocorrências de um padrão de expressão regular por um novo valor em uma string. 

```python
texto = 'meu nome é Felipe'
```

```python
re.sub('[A-Z]','',texto)
```

## Método split

Divide a string pelas ocorrências do  padrão encontrado

```python
texto = 'meu|nome|é|Felipe'
```

```python
re.split('\',texto)
```

```python
re.split('\|',texto)
```

que pode ser reutilizado para buscar correspondências em várias strings.

O uso pode trazer mais velocidade na busca de padrões.

# User-Agents

User-Agent é uma informação presente nos cabeçalhos das solicitações HTTP (Hypertext Transfer Protocol) enviadas pelos clientes (como navegadores da web) para os servidores. Ele fornece aos servidores web detalhes sobre o software e o dispositivo que está fazendo a solicitação.

O User-Agent geralmente contém informações como o nome e a versão do navegador, o sistema operacional, o dispositivo e outras informações relevantes. Por exemplo, um User-Agent típico pode ser "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Safari/537.36". Essa informação permite ao servidor identificar se o cliente está usando o navegador Chrome versão 80.0.3987.163 no Windows 10 de 64 bits.

Como estamos extraindo dados de diferentes fontes , com velocidade e comportamentos diferentes de um usuário convencional, é útil modificar a informação do User-Agents das requisições enviadas via automação ( Python ) , principalmente quando um website bloqueia sua requisição por notar que você não está usando um navegador comum, por exemplo.


```python
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
```

```python
req = requests.get('https://www.uol.com.br',headers=headers)
```

# Uso de Proxy

Um proxy é um servidor intermediário que atua como um intermediário entre um cliente e um servidor. Ele age como uma espécie de "representante" do cliente, enviando as solicitações do cliente ao servidor e recebendo as respostas do servidor de volta ao cliente.

Quando um cliente envia uma solicitação para acessar um recurso na internet, como um site ou um arquivo, essa solicitação passa pelo proxy em vez de ser enviada diretamente ao servidor de destino. O proxy então encaminha a solicitação ao servidor e recebe a resposta, que é enviada de volta ao cliente. Dessa forma, o proxy mascara a identidade e o endereço IP do cliente original para o servidor de destino.

Os proxies podem ser usados por várias razões, como melhorar o desempenho, filtrar o tráfego, contornar restrições de segurança ou acesso geográfico.

Em extração de dados, os proxies são usados para garantir o anonimato e também para diminuir as chances das requisições serem bloqueadas pelo servidor de onde os dados estão sendo extraidos, além de possibilitar acesso a websites que tem restrição por localização geográfica.

# Extraindo dados de PDFs

## Como fazer download de um PDF via requests 

```python

def download_file(url):
    
    import requests 
    
    local_filename = url.split('/')[-1]
    
    with requests.get(url) as r:
        with open(local_filename, 'wb') as f:
            f.write(r.content)
        
    return local_filename
```

```python
download_file('https://bnccompras.blob.core.windows.net/processreports/VencedoresProcessoAdjudicacao20230619090959666.pdf')
```

## Já com o PDF em mãos, vamos extrair o conteúdo

```python
#https://bnccompras.com/Process/ProcessSearchPublic?param1=1
```

```python
import pdfplumber
```

```python
texto = ''

with pdfplumber.open('VencedoresProcessoAdjudicacao20230619090959666.pdf') as pdf:
    for page in pdf.pages:
        texto += page.extract_text()
```

```python
texto.split('\n')
```

```python
import re 
```

```python
cnpj_regex = re.compile(r'[0-9]{2}\.?[0-9]{3}\.?[0-9]{3}\/?[0-9]{4}\-?[0-9]{2} (\d+\.\d+\.?\d+,\d+|\d+,\d+)')
```

```python
for num, row in enumerate(texto.split('\n')):
    
    print([num,row])
```

```python
itens = []  
lista_fornecedores = []    
lotes = []
detalhe_itens = []
descricao_itens = []


for row,line in enumerate(texto.split('\n')):
    if cnpj_regex.search(line):
        if not re.compile(r'LOTE \d+').search(texto.split('\n')[row + 1]):
            if re.compile(r'\s*%s\s*' % 'Gerado em: ').search(texto.split('\n')[row + 1]):
                lista_fornecedores.append([row,line])
            elif not re.compile(r'\s*%s\s*' % 'Gerado em: ').search(texto.split('\n')[row + 1]):
                lista_fornecedores.append([row, line + ' ' + texto.split('\n')[row + 1]])
        else:
            lista_fornecedores.append([row,line])

for row,line in enumerate(texto.split('\n')):
    if re.compile(r'Descrição:').search(line):
        if not re.compile(r'\s*%s\s*' % 'Gerado em: ').search(texto.split('\n')[row + 1]):
            if not re.compile(r'Quantidade: \d+').search(texto.split('\n')[row + 1]):
                row_inside = row
                while not re.compile(r'Quantidade: \d+').search(texto.split('\n')[row_inside + 1]):
                    
                    if re.compile(r'\s*%s\s*' % 'Gerado em: ').search(texto.split('\n')[row_inside + 1]):
                        break
                        
                    row_inside+=1

                descricao_itens.append([row,texto.split('\n')[row:row_inside+1]])

            else:
                descricao_itens.append([row,[line]])
        else:
            descricao_itens.append([row,[line]])
            
            
for row,line in enumerate(texto.split('\n')):

    if re.compile(r'LOTE \d+').search(line):
        lotes.append([row,line])


for row,line in enumerate(texto.split('\n')):
    if re.compile(r'Item: \d+ Unidade:').search(line):
        itens.append([row,line])


for row,line in enumerate(texto.split('\n')):
    if re.compile(r'Quantidade: \d+').search(line):
        detalhe_itens.append([row,line])


dados_itens_vencedores = {}

df_dados_itens_vencedores = pd.DataFrame()

for num,item in enumerate(itens):


    dados_itens_vencedores['lote_item'] =[lote[1] for lote in lotes if lote[0] < item[0]][-1].split('Quant.')[0].strip()

    dados_itens_vencedores['valor_total_lote'] =[lote[1] for lote in lotes if lote[0] < item[0]][-1].split('Total: ')[-1]

    dados_itens_vencedores['descricao_item']  = ' '.join(descricao_itens[num][1])

    dados_itens_vencedores['codigo_item'] = item[1].split('Unidade: ')[0].strip().split('Item: ')[1]

    try:
        dados_itens_vencedores['unidade'] = item[1].split('Unidade: ')[1].split('Marca: ')[0].strip()
    except:
        dados_itens_vencedores['unidade'] = ''

    try:
        dados_itens_vencedores['marca'] = item[1].split('Unidade: ')[1].split('Marca: ')[1].split('Modelo: ')[0].strip()
    except:
        dados_itens_vencedores['marca'] = ''

    try:
        dados_itens_vencedores['modelo'] = item[1].split('Unidade: ')[1].split('Marca: ')[1].split('Modelo: ')[1].strip()
    except:
        dados_itens_vencedores['modelo'] = ''
    try:
        dados_itens_vencedores['quantidade'] = detalhe_itens[num][1].split('Quantidade: ')[1].split('Valor Unit.: ')[0].strip()
    except:
        dados_itens_vencedores['quantidade'] = ''

    try:
        dados_itens_vencedores['valor_unit'] = detalhe_itens[num][1].split('Quantidade: ')[1].split('Valor Unit.: ')[1].split('Total Item: ')[0].strip()
    except:
        dados_itens_vencedores['valor_unit'] = ''

    try:
        dados_itens_vencedores['valor_total'] = detalhe_itens[num][1].split('Quantidade: ')[1].split('Valor Unit.: ')[1].split('Total Item: ')[1].strip()
    except:
        dados_itens_vencedores['valor_total'] = ''

    try:
        dados_itens_vencedores['cnpj_fornecedor'] = re.compile(r'[0-9]{2}\.?[0-9]{3}\.?[0-9]{3}\/?[0-9]{4}\-?[0-9]{2}').search( [fornecedor[1] for fornecedor in lista_fornecedores if fornecedor[0] < [lote[0] for lote in lotes if dados_itens_vencedores['lote_item'] + ' ' in lote[1]][0] ][-1]).group(0)
    except:
        dados_itens_vencedores['cnpj_fornecedor'] = ''

    try:
        dados_itens_vencedores['razao_social_fornecedor'] = " ".join([re.sub('\d+\.\d+,\d+|\d+,\d+','',re.sub(r'[0-9]{2}\.?[0-9]{3}\.?[0-9]{3}\/?[0-9]{4}\-?[0-9]{2}','',fornecedor[1])) for fornecedor in lista_fornecedores if fornecedor[0] < [lote[0] for lote in lotes if dados_itens_vencedores['lote_item'] + ' ' in lote[1]][0] ][-1].split()) 
    except:
        dados_itens_vencedores['razao_social_fornecedor'] = ''





    df_in_itens = pd.DataFrame([dados_itens_vencedores])
    df_dados_itens_vencedores = pd.concat([df_dados_itens_vencedores, df_in_itens], ignore_index=True)
```

```python
import requests 
from bs4 import BeautifulSoup 
 
login = LOGIN_GITHUB 
password =  SENHA_GITHUT
login_url = "https://github.com/session" 
repos_url = "https://github.com/" + login + "/?tab=repositories" 
 
with requests.session() as s: 
    req = s.get(login_url).text 
    html = BeautifulSoup(req,"html.parser") 
    token = html.find("input", {"name": "authenticity_token"}).attrs["value"] 
    time = html.find("input", {"name": "timestamp"}).attrs["value"] 
    timeSecret = html.find("input", {"name": "timestamp_secret"}).attrs["value"] 

    payload = { 
        "authenticity_token": token, 
        "login": login, 
        "password": password, 
        "timestamp": time, 
        "timestamp_secret": timeSecret 
    } 
    res =s.post(login_url, data=payload) 

    r = s.get(repos_url) 
    soup = BeautifulSoup (r.content, "html.parser") 
    usernameDiv = soup.find("span", class_="p-nickname vcard-username d-block") 
    print("Username: " + usernameDiv.getText()) 

    repos = soup.find_all("h3", class_="wb-break-all") 
    for r in repos: 
        repoName = r.find("a").getText() 
        print("Repository Name: " + repoName)
```

# Paralelisando Extrações

```python
import concurrent.futures
```

```python
concurrent.futures.ThreadPoolExecutor
```

```python

# recommended is equal to the number of CPU Cores.
max_workers = 2
```

```python
df_final = pd.DataFrame()
list_errors_final = []

import time 


time_init = time.time()


with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
    future_to_url= (executor.submit(captura_medicina,url) for url in lista_urls[0:30])

    for cont,future in enumerate(concurrent.futures.as_completed(future_to_url)):
    
        try:
        
            if type(future.result()) != str:
        
                df_final_in = pd.DataFrame(future.result())
                
 
            
                df_final = pd.concat([df_final_in, df_final])
            
            else:
                
                list_errors_final.append(future.result())
        
            print("Extraindo > ",cont+1 ,"de " ,len(lista_urls))
            
        except Exception as E:
            print(E)
            pass

time_fim = time.time()

print(time_fim - time_init , ' segundos ')
```

```python
lista_urls= []
for url in range(0,319):
    
    lista_urls.append(f"https://www.1mg.com/pharmacy_api_gateway/v4/drug_skus/by_prefix?prefix_term=b&page={url}&per_page=30")
```

```python
def captura_medicina(url):
    
    req = requests.get(url)
    
    return pd.json_normalize(pd.json_normalize(req.json())['data.skus'][0])
```
